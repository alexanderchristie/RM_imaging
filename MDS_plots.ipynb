```python
#MDS class and methods

class MDS_obj():
    def __init__(self, X, n_neighbors=4, Radius=.65,dim=3, EUC=True):
        super(MDS_obj, self).__init__()

        torch_inner=np.abs(1.0-(np.abs(np.inner(X,X.conj()))))
        print(torch_inner.shape)
        neigh = NearestNeighbors(n_neighbors=n_neighbors, radius=Radius, metric='precomputed')
        neigh.fit(torch_inner)
        if n_neighbors>1:
            neighbors_dist, neighbors=neigh.kneighbors()
        else:
            neighbors_dist, neighbors=neigh.radius_neighbors()
        dist_matrix=np.zeros((torch_inner.shape[0],torch_inner.shape[0]))
        for i in range(torch_inner.shape[0]):
            for j in range(len(neighbors[i])):
                if not EUC:
                    dist_matrix[i][neighbors[i][j]]=1
                    dist_matrix[neighbors[i][j]][i]=1#neighbors_dist[i][j]
                else:
                    dist_matrix[i][neighbors[i][j]]=neighbors_dist[i][j]
                    dist_matrix[neighbors[i][j]][i]=neighbors_dist[i][j]
        A=dist_matrix
        avg_neighbors=sum([len(A[i][A[i]>0]) for i in range(torch_inner.shape[0])])/torch_inner.shape[0]
        
        print(f'average number of neighbors in single case: {avg_neighbors}')
        A=shortest_path(A)
        
        self.D=A
        

        self.proxy_mds(A, dim=dim)
        
        
    

    def proxy_mds(self, D, dim=3):
        L=np.eye(D.shape[0])-1/D.shape[0]*np.ones((D.shape[0],D.shape[0]))
        M=L@D@L
        eigvals, eigvecs=np.linalg.eig(M)
        eigvals=np.real(eigvals)
        eigvecs=np.real(eigvecs)

        eigvecs=eigvecs[:,np.argsort(eigvals)]
        eigvals=np.sort(eigvals)

        
        U, S, Vh = np.linalg.svd(M)
        self.sing_vals=copy.deepcopy(S)




        eigvals=eigvals[::-1]
        eigvecs=eigvecs[::-1]


        self.eig_vals=copy.deepcopy(eigvals)
        self.eig_vects=copy.deepcopy(eigvecs)
      




        eigvecs=eigvecs[:,:dim]
        eigvals=eigvals[:dim]
        eigvals=np.diag(np.sqrt(eigvals))
        eigvecs=eigvecs@eigvals
        self.X=eigvecs



#IMAGES OF INNER PRODUCTS ACROSS 2 different random media
def plot_2_imgs(inners, rho_hat,J,ind=9,figsize=8,scaling='Linf',font_size=50, Single=False, xpix=31, ypix=21, file_name=None, Lclim=None, Hclim=None):
   # plt.close()
    rho_hat_1=np.zeros((2, 361))
    rho_hat_1[0, :]=inners[J,:361]
    
    #rho_hat_1=torch.tensor(rho_hat_1)


    
    rho_hat_2=np.zeros((2, 361))
    rho_hat_2[0, :]=inners[J,361:]
    #rho=torch.tensor(rho)
    #rho_hat_2=torch.tensor(rho_hat_2)




    figsize=(figsize*2,figsize)
    fig, axes=plt.subplots(nrows= 1, ncols= 2,figsize=figsize)
    rhos=[rho_hat_1,rho_hat_2]
    for i in [0,1]:
        rho_hat=rhos[i]
        tick_params = {'labelsize': font_size}
        output=rho_hat.squeeze() 
        output=output[ind,:]
        #img=torch.abs((output.squeeze())).reshape(xpix, ypix, order='F')
        ax=axes[i]
        img=np.abs(output).reshape(19,19, order='F')
        ax.pcolor(np.abs(output).reshape(19,19, order='F'),  cmap=parula_map)

        
        pcol2=ax.pcolor(np.abs(output).reshape(19,19, order='F'),  cmap=parula_map)
        cbar=plt.colorbar(pcol2, ax=ax)
    #plt.title('true', fontsize=font_size)
        #ax.set_title('Full model', fontsize=font_size)
        cbar.mappable.set_clim(Lclim, Hclim)
        
        cbar.ax.tick_params(labelsize=font_size)

        #cbar.remove()
        ax.tick_params(axis='both', **tick_params)
        ax.set_xticks( ticks=[1,4,7,10,13,16,19], minor=False)#, font_size=font_size)
        ax.set_yticks( ticks=[1,4,7,10,13,16,19], minor=False)#, font_size=font_size)
        
        #if i==0:
        #cbar.remove()
    if file_name:
        plt.savefig(f'/home/achristie/Codes_data/Experiment_data/library_learning/{file_name}.pdf')
    plt.show()



def nbdhplot(med1=0, med2=1, indices=[10*19+10]):
    data_path=os.path.join(cwd,f'Data/FoldyLox_all_seeds/FoldyLox_seed{med1}')
    medium1= np.array(mat73.loadmat(data_path+'/rtt.mat')['Artt'])
    data_path=os.path.join(cwd,f'Data/FoldyLox_all_seeds/FoldyLox_seed{med2}')
    medium2= np.array(mat73.loadmat(data_path+'/rtt.mat')['Artt'])
    medium=np.concatenate((medium1.T, medium2.T), axis=0)
    medium=medium.T

    print(medium.shape)


    inners=medium.transpose().conjugate()@medium
    inners=np.abs(inners)




    coherence=0
    for i in range(361):
        for j in range(361, 722):
            if inners[i][j]>coherence and i!=j:
                coherence=inners[i][j]
    print(f'Max_(i, j)<g_1, i, g_2, j>={coherence}')

    coherence=0
    for i in range(361):
        if inners[i][i+361]>coherence and i!=j:
            coherence=inners[i][i+361]
    print(f'Max_(i)<g_1, i, g_2, i>={coherence}')


    coherence=0
    for i in range(361):
        for j in range(361):
            if inners[i][j]>coherence and i!=j:
                coherence=inners[i][j]
    print(f'Max_(i, j)<g_1, i, g_1, j>={coherence}')

    coherence=0
    for i in range(361,361*2):
        for j in range(361,361*2):
            if inners[i][j]>coherence and i!=j:
                coherence=inners[i][j]
    print(f'Max_(i, j)<g_2, i, g_2, j>={coherence}')


    coherence=0
    for i in range(len(inners)):
        for j in range(len(inners[i])):
            if inners[i][j]>coherence and i!=j and j!=i+361 and j!=i-361:
                coherence=inners[i][j]
    print(f'Max_(i, j!=i)<g_1, i, g_,2j>={coherence}')





    neigh = NearestNeighbors(n_neighbors=1, radius=0.00001, metric='precomputed')
    col_list=medium.T
    torch_inner=np.abs(1.0-(np.abs(np.inner(col_list,col_list.conj()))))

    connections=neigh.fit(torch_inner)
    neighbors_dist, neighbors=neigh.kneighbors()
    not_split=0
    if False:
        for i in range(361):
            for j in neighbors[i]:
                if j>360 and j!=i+361:
                    print(i, neighbors[i], [neighbors[i][k]-361 for k in range(4)])
                    not_split+=1
                    #break


    J=10*19+10
    rho_hat=np.zeros((2, 361))
    rho=np.zeros(361)
    rho_hat[0, :]=inners[J,:361]
    #rho=torch.tensor(rho)
    rho_hat=torch.tensor(rho_hat)
    #rho[22+20]=1
    #rho[400-22-1-20]=1
    #rho[400-37-1-20]=1
    #rho[210]=1



    rho[37+20]=1
    rho[22+20]=1
    rho[400-22-1-20]=1
    rho[400-37-1-20]=1
    rho[210]=1

    for J in indices:
        rho_hat=np.zeros((2, 361))

        rho_hat[0, :]=inners[J,:361]
        #rho=torch.tensor(rho)
        plot_2_imgs(inners, rho_hat,J,ind=0, figsize=5, scaling=None, xpix=19, ypix=19, font_size=15, Single=True, Lclim=0, Hclim=1, file_name=None)
    #inners[J,:361]
    #rho_hat[0, :]=inners[J,361:]
    #rho=torch.tensor(rho)
    #rho_hat=torch.tensor(rho_hat)
```

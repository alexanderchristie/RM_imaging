```python
#MDS class.

class MDS_obj():
    def __init__(self, X, n_neighbors=4, Radius=.65,dim=3, EUC=True):
        super(MDS_obj, self).__init__()

        torch_inner=np.abs(1.0-(np.abs(np.inner(X,X.conj()))))
        print(torch_inner.shape)
        neigh = NearestNeighbors(n_neighbors=n_neighbors, radius=Radius, metric='precomputed')
        neigh.fit(torch_inner)
        if n_neighbors>1:
            neighbors_dist, neighbors=neigh.kneighbors()
        else:
            neighbors_dist, neighbors=neigh.radius_neighbors()
        dist_matrix=np.zeros((torch_inner.shape[0],torch_inner.shape[0]))
        for i in range(torch_inner.shape[0]):
            for j in range(len(neighbors[i])):
                if not EUC:
                    dist_matrix[i][neighbors[i][j]]=1
                    dist_matrix[neighbors[i][j]][i]=1#neighbors_dist[i][j]
                else:
                    dist_matrix[i][neighbors[i][j]]=neighbors_dist[i][j]
                    dist_matrix[neighbors[i][j]][i]=neighbors_dist[i][j]
        A=dist_matrix
        avg_neighbors=sum([len(A[i][A[i]>0]) for i in range(torch_inner.shape[0])])/torch_inner.shape[0]
        
        print(f'average number of neighbors in single case: {avg_neighbors}')
        A=shortest_path(A)
        
        self.D=A
        

        self.proxy_mds(A, dim=dim)
        
        
    

    def proxy_mds(self, D, dim=3):
        L=np.eye(D.shape[0])-1/D.shape[0]*np.ones((D.shape[0],D.shape[0]))
        M=L@D@L
        eigvals, eigvecs=np.linalg.eig(M)
        eigvals=np.real(eigvals)
        eigvecs=np.real(eigvecs)

        eigvecs=eigvecs[:,np.argsort(eigvals)]
        eigvals=np.sort(eigvals)

        
        U, S, Vh = np.linalg.svd(M)
        self.sing_vals=copy.deepcopy(S)




        eigvals=eigvals[::-1]
        eigvecs=eigvecs[::-1]


        self.eig_vals=copy.deepcopy(eigvals)
        self.eig_vects=copy.deepcopy(eigvecs)
      




        eigvecs=eigvecs[:,:dim]
        eigvals=eigvals[:dim]
        eigvals=np.diag(np.sqrt(eigvals))
        eigvecs=eigvecs@eigvals
        self.X=eigvecs





```
